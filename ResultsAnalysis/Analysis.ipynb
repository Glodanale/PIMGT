# Results Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Aggregation.csv")
```

```python
# Architecture-based DataFrames
df_mgt = df[[col for col in df.columns if col.startswith("MGT_")]]
df_pimgt_ex = df[[col for col in df.columns if col.startswith("PIMGT_EX_")]]
df_pimgt_in = df[[col for col in df.columns if col.startswith("PIMGT_IN_")]]

# Seed-based DataFrames
df_seed25 = df[[col for col in df.columns if col.endswith("seed25")]]
df_seed42 = df[[col for col in df.columns if col.endswith("seed42")]]
df_seed79 = df[[col for col in df.columns if col.endswith("seed79")]]
```

```python
print(df_mgt)
print(df_pimgt_ex)
print(df_pimgt_in)
print(df_seed25)
print(df_seed42)
print(df_seed79)
```

```python
def plot_rmse_lines(df, title="RMSE Over Time", figsize=(10, 6)):
    """
    Plots all RMSE columns in the given dataframe as individual lines.
    
    Parameters:
    - df: DataFrame containing RMSE columns (column names must include 'rmse')
    - title: Title of the plot
    - figsize: Size of the matplotlib figure
    """
    plt.figure(figsize=figsize)
    
    # Plot only columns containing 'rmse'
    for col in df.columns:
        if 'rmse' in col.lower():
            plt.plot(df.index, df[col], label=col)

    plt.xlabel("Prediction Step")
    plt.ylabel("RMSE")
    plt.title(title)
    plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1), borderaxespad=0)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{title}.png", bbox_inches='tight')
    plt.show()
```

```python
def plot_mae_lines(df, title="MAE Over Time", figsize=(10, 6)):
    """
    Plots all MAE columns in the given dataframe as individual lines.
    
    Parameters:
    - df: DataFrame containing MAE columns (column names must include 'mae')
    - title: Title of the plot
    - figsize: Size of the matplotlib figure
    """
    plt.figure(figsize=figsize)
    
    for col in df.columns:
        if 'mae' in col.lower():
            plt.plot(df.index, df[col], label=col)

    plt.xlabel("Prediction Step")
    plt.ylabel("MAE")
    plt.title(title)
    plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1), borderaxespad=0)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{title}.png", bbox_inches='tight')
    plt.show()
```

```python
def plot_mape_lines(df, title="MAPE Over Time", figsize=(10, 6)):
    """
    Plots all MAPE columns in the given dataframe as individual lines.
    
    Parameters:
    - df: DataFrame containing MAPE columns (column names must include 'mape')
    - title: Title of the plot
    - figsize: Size of the matplotlib figure
    """
    plt.figure(figsize=figsize)
    
    for col in df.columns:
        if 'mape' in col.lower():
            plt.plot(df.index, df[col], label=col)

    plt.xlabel("Prediction Step")
    plt.ylabel("MAPE (%)")
    plt.title(title)
    plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1), borderaxespad=0)
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{title}.png", bbox_inches='tight')
    plt.show()
```

```python
plot_rmse_lines(df_mgt, title="MGT Architecture - RMSE")
plot_rmse_lines(df_pimgt_ex, title="PIMGT External - RMSE")
plot_rmse_lines(df_pimgt_in, title="PIMGT Internal - RMSE")
plot_rmse_lines(df_seed25, title="Seed 25 - RMSE Across Architectures")
plot_rmse_lines(df_seed42, title="Seed 42 - RMSE Across Architectures")
plot_rmse_lines(df_seed79, title="Seed 79 - RMSE Across Architectures")
```

```python
plot_mae_lines(df_mgt, title="MGT Architecture - MAE")
plot_mae_lines(df_pimgt_ex, title="PIMGT External - MAE")
plot_mae_lines(df_pimgt_in, title="PIMGT Internal - MAE")
plot_mae_lines(df_seed25, title="Seed 25 - MAE Across Architectures")
plot_mae_lines(df_seed42, title="Seed 42 - MAE Across Architectures")
plot_mae_lines(df_seed79, title="Seed 79 - MAE Across Architectures")
```

```python
plot_mape_lines(df_mgt, title="MGT Architecture - MAPE")
plot_mape_lines(df_pimgt_ex, title="PIMGT External - MAPE")
plot_mape_lines(df_pimgt_in, title="PIMGT Internal - MAPE")
plot_mape_lines(df_seed25, title="Seed 25 - MAPE Across Architectures")
plot_mape_lines(df_seed42, title="Seed 42 - MAPE Across Architectures")
plot_mape_lines(df_seed79, title="Seed 79 - MAPE Across Architectures")
```
